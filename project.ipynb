```python
f1 = ['free','money','secure','restore','fund','service']
f2 = ['failure','irregular','locked','incorrect','invalid','fraud','blocked','restricted','limited']
f3 = ['password','holder','payment','attempts','banking','fraudulant','copyright','prevent', 'access']
f4 = ['cost','tax','technology','letter','limit','execute','exchange','draft','agreement','legal','action']
def Load_data():
  text,w2vgain,w2loss,w2gnls,w2none,labels = [], [], [], [], [], []
  with open("datatest2.txt",encoding='utf-8-sig') as f: #loads dataset python 3, encodinf stig stops dataset breaking
    for line in f:
      split = line.split()
      labels.append(split[0].strip())
      text.append(''.join(split[1:]).strip())
  for sentence in text:
     if (any(map(lambda word: word in sentence, f1))):
        print(sentence)
        w2vgain.append(sentence)
  for sentence in text:
     if (any(map(lambda word: word in sentence, f2))):
        print(sentence)
        w2loss.append(sentence)
  for sentence in text:
     if (any(map(lambda word: word in sentence, f3))):
        print(sentence)
        w2gnls.append(sentence)
  for sentence in text:
     if (any(map(lambda word: word in sentence, f4))):
        print(sentence)
        w2none.append(sentence)
  return text,w2vgain,w2loss,w2gnls,w2none,labels
```

```python
def review_to_words(raw_review):

    r = re.sub("[^a-zA-Z]", " ", raw_review)

    r = r.lower()

    words = r.split()


    words = [w for w in words if not w in stopwords.words("english")]

    return(" ".join(words))


    # clean all reviews in an array
```

```python
def clean_all(X):

    num_reviews = len(X)

    clean_reviews = []


    for i in range(0, num_reviews):

        if( (i+1)%100 == 0 ):

            print("Cleaned %d/%d Reviews" % ( i+1, num_reviews ))

        clean_reviews.append(review_to_words(X[i]))
    return(clean_reviews)
```
